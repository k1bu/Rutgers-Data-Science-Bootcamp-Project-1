```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file = "checkouts_5yr.csv"

data = pd.read_csv(file)
```

```python
gb_MatType = data.groupby(['MaterialType', 'CheckoutYear', 'Title'])
funcs = {'Checkouts':'sum'}
test_agg = gb_MatType.agg(funcs)

test_agg
```

```python
test2_agg = test_agg.sort_values(['MaterialType', 'CheckoutYear', 'Checkouts'], ascending = False)
test2_agg
```
