```python
import os
import csv
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from datetime import datetime
import matplotlib as mpl



%matplotlib inline

file = "seattle.csv"



seattle_w = pd.read_csv(file)

seattle_w.head()
```

```python
# adding month and year columns

seattle_w['year'] = pd.DatetimeIndex(seattle_w['date']).year
seattle_w['month'] = pd.DatetimeIndex(seattle_w['date']).month

seattle_w.head()
```

```python
seattle = seattle_w.groupby(["year","month"])
funcs = {"temp":"mean"}

months = seattle.agg(funcs)

months

```

```python
# groupby

seattle_month = seattle_w.groupby(["month"])
funcs = {"temp":"mean"}

months_year = seattle_month.agg(funcs)

months_year
```

```python
# plotting the average monthy temp in Seattle (all years combined)

months_year.plot(kind='line',title='Avgerage Monthly Temperature In Seattle')
plt.style.use('ggplot')
plt.xlabel('Month')
plt.ylabel('Average Monthly Temperature')

print(months_year)

```

```python
file = 'C:\\Users\\hjp28\\Desktop\\Group Project 1\\checkouts_5yr.csv'
df = pd.read_csv(file)

df.head()
```

```python
# Using code from Veeraj for Avg. Number of Checkouts per Month

df.count()

df['MaterialType'].value_counts()

df["CheckoutYear"].value_counts()


df["CheckoutMonth"].value_counts()


groupby_title = df.groupby(["Title"])
groupby_title.head()

Total_Checkouts_per_Title = groupby_title["Checkouts"].count()
Total_Checkouts_per_Title.head()

gb_title = df.groupby(['Title'])
funcs = {'Checkouts':'sum'}
title_agg = gb_title.agg(funcs)




sort_title = title_agg.sort_values("Checkouts",ascending = False)
sort_title



gb_month = df.groupby(['CheckoutYear','CheckoutMonth'])
funcs = {'Checkouts':'sum'}
month_agg = gb_month.agg(funcs)


month_agg

gb_month_total = month_agg.groupby(['CheckoutMonth'])
funcs = {'Checkouts':'mean'}
all_month_agg = gb_month_total.agg(funcs)

all_month_agg.index.names = ["month"]

all_month_agg





all_month_agg.plot(kind='line',title='Avg. Checkouts per Month')
plt.xlabel('Month')
plt.ylabel('Average Number of Checkouts')

print(all_month_agg)

```

```python
# combining the checkouts and temperature graphs

t = np.arange(1,53,1)

plt.style.use('ggplot')
data1 = months
data2 = month_agg

fig, ax1 = plt.subplots(figsize=(20,10))

plt.title("Average Temperature vs Average Checkouts in Seattle by Month")


color = 'tab:red'
ax1.set_xlabel('Months')
ax1.set_ylabel('Temperature', color=color)
ax1.plot(t, data1, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx() 

color = 'tab:blue'
ax2.set_ylabel('Checkouts', color=color)  
ax2.plot(t, data2, color=color)
ax2.tick_params(axis='y', labelcolor=color)


#fig.tight_layout() 


plt.savefig('tempvscheckouts.png')

plt.show()


```
