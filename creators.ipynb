```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file = "checkouts_5yr.csv"

data = pd.read_csv(file)
```

```python
gb_creators = data.groupby(['Creator'])
funcs = {'Checkouts':'sum'}
creators_agg = gb_creators.agg(funcs)
```

```python
creator_ranked = creators_agg.sort_values('Checkouts',ascending = False)
creator10 = creator_ranked.head(10)
creator10
```

```python
plt.figure(figsize=(50,25))
plt.rcParams.update({'font.size': 32})
x = ["Mo Willems", "Geronimo Stilton", "Jim Davis", "Daisy Meadows", "Dr. Seuss", "James Patterson", "Mary Pope Osborne", "Tedd Arnold", "Jennifer Holm", "Dav Pilkey"]
checkouts = creator10.loc[:,"Checkouts"]
x_pos = [i for i, _ in enumerate(x)]
plt.barh(x_pos, checkouts)
plt.yticks(x_pos,x)
plt.gca().invert_yaxis()
plt.xlabel('Checkouts', labelpad = 40)
plt.ylabel('Creator', labelpad = 40)
plt.title('Top 10 Creators by Checkout', fontsize = 80, pad = 40)
fig1 = plt.gcf()
fig1.savefig('topcreators.png')
```
