```python
# Dependencies
import requests
import json

# Google developer API key
from config import gkey

```

```python
target_coordinates = "47.6038,-122.3301"
target_search = "The Seattle Public Library"
target_radius = 50000
target_type = "public library"

# set up a parameters dictionary
params = {
    "location": target_coordinates,
    "keyword": target_search,
    "radius": target_radius,
    "type": target_type,
    "key": gkey
}

# base url
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

# run a request using our params dictionary
response = requests.get(base_url, params=params)
```

```python

print(response)


```

```python
# convert response to json
places_data = response.json()

# Print the json (pretty printed)
print(json.dumps(places_data, indent=4, sort_keys=True))


```

```python
# Print the name and address of the first restaurant that appears
print(places_data["results"][2]["name"])
print(places_data["results"][2]["vicinity"])
print(places_data["results"][2]["geometry"])
```
